## Big O

### O(n)

```python
# O(n) - Linear Time Complexity
# This function demonstrates O(n) time complexity, which means the execution time
# grows linearly with the input size. In this example, we have a single loop that
# runs exactly 'n' times, where 'n' is the input parameter. If n = 10, the loop
# executes 10 times; if n = 100, it executes 100 times. The relationship between
# input size and number of operations is directly proportional (1:1 ratio).
# This is considered efficient for algorithms that must process every element once.

def print_items(n):
    for i in range(n):
        print(i)

print_items(10)

```

### O(n²)

```python
# O(n²) - Quadratic Time Complexity
# This function demonstrates O(n²) time complexity, which means the execution time
# grows quadratically with the input size. We have nested loops where the outer loop
# runs 'n' times and for each iteration of the outer loop, the inner loop also runs
# 'n' times. This results in n × n = n² total operations. If n = 10, we get 100
# operations; if n = 100, we get 10,000 operations. The growth rate is exponential
# compared to O(n), making it less efficient for large datasets but common in
# algorithms that compare every element with every other element.

def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i, j)

print_items(10)
```

### O(1)

```python
# O(1) - Constant Time Complexity
# This function demonstrates O(1) time complexity, which means the execution time
# remains constant regardless of the input size. Even though we're performing
# three addition operations (n + n + n), the number of operations doesn't change
# based on the value of 'n'. Whether n = 10 or n = 1,000,000, we always perform
# exactly 3 operations. This is the most efficient time complexity because the
# algorithm's performance doesn't degrade as the input size grows. Examples include
# accessing an array element by index, simple arithmetic operations, or returning a value.

def print_items(n):
    return n + n + n

print(print_items(10))
```

### O(Log n)

```python
# O(log n) - Logarithmic Time Complexity
# This function demonstrates O(log n) time complexity, which means the execution time
# increases logarithmically as the input size grows. Typically, algorithms that
# repeatedly divide the problem in half (like binary search) have O(log n) complexity.
# Here, we use a loop that halves 'n' each time, printing the current value of 'n'
# until it becomes less than or equal to 1.

def print_items(n):
    while n > 1:
        print(n)
        n = n // 2

print_items(16)

```

```python
# Big O Complexity Visualization
# This plot shows how different time complexities grow as input size increases.
# Notice how O(1) stays flat, O(log n) grows slowly, O(n) grows linearly,
# O(n log n) grows faster than linear, and O(n²) grows exponentially.

import matplotlib.pyplot as plt
import numpy as np

# Generate input sizes from 1 to 100
n = np.arange(1, 101)

# Calculate different time complexities
o_1 = np.ones_like(n)  # O(1) - constant
o_log_n = np.log2(n)   # O(log n) - logarithmic
o_n = n                # O(n) - linear
o_n_log_n = n * np.log2(n)  # O(n log n) - linearithmic
o_n_squared = n ** 2   # O(n²) - quadratic

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(n, o_1, label='O(1) - Constant', linewidth=2, color='green')
plt.plot(n, o_log_n, label='O(log n) - Logarithmic', linewidth=2, color='blue')
plt.plot(n, o_n, label='O(n) - Linear', linewidth=2, color='orange')
plt.plot(n, o_n_log_n, label='O(n log n) - Linearithmic', linewidth=2, color='red')
plt.plot(n, o_n_squared, label='O(n²) - Quadratic', linewidth=2, color='purple')

# Customize the plot
plt.xlabel('Input Size (n)', fontsize=12)
plt.ylabel('Number of Operations', fontsize=12)
plt.title('Big O Time Complexity Comparison', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(1, 100)
plt.ylim(0, 1000)

# Add annotations to highlight key differences
plt.annotate('Most Efficient', xy=(50, 1), xytext=(60, 100),
            arrowprops=dict(arrowstyle='->', color='green', alpha=0.7),
            fontsize=10, color='green')
plt.annotate('Least Efficient\n(for large n)', xy=(80, 6400), xytext=(40, 8000),
            arrowprops=dict(arrowstyle='->', color='purple', alpha=0.7),
            fontsize=10, color='purple')

plt.tight_layout()
plt.show()
```

### O(n log n)

```python
# O(n log n) - Linearithmic Time Complexity
# This function demonstrates O(n log n) time complexity, which is common in efficient sorting algorithms
# like mergesort and heapsort. The function contains a loop that runs 'n' times, and inside that loop,
# another loop runs log(n) times (by halving 'j' each iteration). The total number of operations is
# proportional to n * log n.

def print_items(n):
    i = 0
    while i < n:
        j = n
        while j > 1:
            print(f"i={i}, j={j}")
            j = j // 2
        i += 1

print_items(8)
```
